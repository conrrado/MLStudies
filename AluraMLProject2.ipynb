```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/b9dd8e4b62b9e22ebcb9c8e89c271de4/raw/c69ec4b708fba03c445397b6a361db4345c83d7a/tracking.csv"
data = pd.read_csv(uri)
data.head()
```

```python
y = data["comprou"]
y.head()
```

```python
x = data[["inicial", "palestras", "contato", "patrocinio"]]
x.head()
```

```python
# line x column
data.shape
```

```python
# Only the first 75 items will be used for training, the remaining items will be
#   used for testing the model
x_train = x[:75]
y_train = y[:75]

x_test = x[75:]
y_test = y[75:]

print(f"Let's train with {len(x_train)} itens")
print(f"Let's test with {len(x_test)} itens")
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

model = LinearSVC()
model.fit(x_train, y_train)
predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions) * 100
print(f"Accuracy rate: {accuracy:.2f}%")
```

# Using the library to separate training and testing

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

SEED = 20

# random_state -> ensure that the division of data always the same when run this
#   code again
# stratify -> ensure the same proportion in training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=SEED, stratify=y)

print(f"Let's train with {len(x_train)} itens")
print(f"Let's test with {len(x_test)} itens")

model = LinearSVC()
model.fit(x_train, y_train)
predictions = model.predict(x_test)

accuracy = accuracy_score(y_test, predictions) * 100
print(f"Accuracy rate: {accuracy:.2f}%")
```
